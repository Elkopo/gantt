```python
# imports 
import pandas as pd
import numpy as np
import os
from datetime import datetime, timedelta
import itertools
```

```python
# read formatted excel -- adjust/add code to format UAPR schedules
df = pd.read_excel('008_sched.xlsx')
df
```

```python
# combine date and time series into one list
temp_dtg = []
for index, row in df.iterrows():
  temp_dtg.append(pd.Timestamp.combine(row['Date'], row['Start_Time']))
```

```python
# reset Date series and set index to DTG
df['Date'] = pd.Series(temp_dtg)
df.set_index('Date', inplace=True)
```

```python
# set timezone to local -- change as needed
df = df.tz_localize(tz='US/Arizona')
```

```python
# drop uncessary rows
df = df[df['Lesson_Activity'] != 'Lunch']
df
```

```python
df.index[0]
```

```python
c.events.clear()
c
```

```python
# add dataframe events to ics calendar
for index, row in df.iterrows():
  c.events.add(Event(
      name=row['Description'],
      begin=str(row.name),
      duration=pd.to_timedelta(str(row['Hours'])),
      location=str(row['Resources']),
      organizer=str(row['Trainers']),
      categories=str(row['Lesson_Activity'])))
```
